```python
# start dialation and erosion 

# path to image
path = r'../get_graph_of_zig_zag/zigzag_full.png'
image = open_image(path)

show_image(image)
option = 2 # 1 = erode first 2 = dialate first 
num_erosions = 0
num_dilations = 0

# inputs = path to image, erode first (1) or dialate first (2) .desired number of erosions, desired number of dialations 
# outputs = eroded then dialated image
# e_d_image , name = erosion_dilation_from_array(path,option,num_erosions,num_dilations)
e_d_image  = erosion_dilation_from_array(image,option,num_erosions,num_dilations)


image = e_d_image

# print(erosion_dilation(path,option,num_erosions,num_dilations))

show_image(image)

```

```python
#skeleton to graph

# image = open_image(path)

# print(image)

graph = skeleton_to_graph_from_array(image)

print(type(graph)) #<class 'networkx.classes.graph.Graph'>

# draw image
for (s,e) in graph.edges():
    ps = graph[s][e]['pts']
    plt.plot(ps[:,1], ps[:,0], 'green')
    
# draw node by o
nodes = graph.nodes()
ps = np.array([nodes[i]['o'] for i in nodes])
plt.plot(ps[:,1], ps[:,0], 'r.')
i = 0
for x,y in zip(ps[:,1],ps[:,0]):
    
    label = str(i)
    
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
    break
    i=i+1


plt.imshow(image)
# title and show
plt.title('Build Graph')


plt.savefig("graph.png")


plt.show()

```
